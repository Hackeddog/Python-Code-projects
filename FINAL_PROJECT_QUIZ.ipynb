```python
import random


questionnaire_type = ("word_bank","multiple_choice")

class General:
    def __init__(self, filename):
        self.filename = filename
        self.dictionary = {}
        self.answers = []
    def prepare_questions(self):
        self.questions = shuffle_questions(self.dictionary)
    def read_file(self):
        try: #error handling
            with open(self.filename, "r") as file:
                for line in file:
                    parts = line.strip().split(",") #reads the line in the file
                    question = parts[0].strip()
                    answer = parts[1].strip()
                    self.dictionary[question] = (answer.upper()) # Adds it to the dictionary
            return True
        except Exception as e:
            print(f"An Error occured: {e} .Please check your file")
            return False


    def show_summary(self, type_of_questionnaire):
      with open(f"summary_{type_of_questionnaire}.txt", "w") as fout:
        text = []
        correct = 0
        i=0
        text.append(f"{'-':-^60}\n")
        text.append(f"{'SUMMARY OF THE TEST':^60}\n")
        text.append(f"{'-':-^60}\n")
        text.append(f"{'NUMBER OF QUESTIONS: '}{len(self.questions):<9}{'TYPE OF TEST: '}{type_of_questionnaire}\n")
        text.append(f"{'-':-^60}\n")
        for (question, correct_answer) in self.questions.items():
          answer = self.answers[i]
          text.append(f"({i+1}) {question}\n")
          text.append(f"Your Answer: {answer}\n")
          text.append(f"Correct Answer: {correct_answer.upper()}\n")
          text.append(f"{'-':-^60}\n")
          i += 1
          if answer == correct_answer.upper():
            correct += 1
        percentage = (correct/len(self.questions))*100
        text.append(f"{'-':-^60}\n")
        text.append(f"{'Correct Answers: '}{correct}/{len(self.questions)}\n")
        text.append(f"Percentage: {percentage:.2f}% \n")
        text.append(f"{'-':-^60}\n")
        fout.writelines(text)


#----WORD BANK CLASS--------------------------------------------------------------------
class WordBank(General):
    def start(self, type_of_questionnaire):
        self.general = General("question_test.txt")  # Connecting the General instance
        self.general.read_file()  # Ensure questions are read from the file
        self.general.prepare_questions() # Shuffle the questions
        line()
        print(f"{'WORD BANK':^90}")
        line()
        self.word_bank_box()  # Display the choices
        line()
        print(f"{'Choose the answers from the above'}")
        line()
        i = 1
        for question in self.general.questions.keys():  # iterate each questions
            while True:
                answer = input(f"({i}) {question}: ").upper()
                if answer in self.general.questions.values():
                    print(f"{'Your Answer:':>15}{answer}")
                    self.general.answers.append(answer) # records the answer of the user
                    line()
                    i += 1
                    break  # Exit inner loop if answer is correct
                else:
                    print(f"ANSWER NOT IN THE CHOICES ABOVE TRY AGAIN")
                    line()
        self.general.show_summary(type_of_questionnaire)
        line()
        print(f"{'YOU HAVE ANSWERED ALL OF THE QUESTIONS':^90}")
        line()

    def word_bank_box(self):
        self.choices = shuffle_questions(self.general.questions)  # Use the shuffled questions from General
        line()
        counter = 0
        for answer in (self.choices.values()):
            print(f"{answer:<20}", end=" ")
            counter += 1
            if counter == 4:  # Move to next line after 4 answers
                print()
                counter = 0
        print()

#-----MULTIPLE CHOICE CLASS-------------------------------------------------------
class MultipleChoice(General):
    def start(self, type_of_questionnaire):
        choices =('A','B','C','D')
        self.general = General("question_test.txt")  # Connecting the General instance
        self.general.read_file()  # Ensure questions are read from the file
        self.general.prepare_questions() # Shuffle the questions
        line()
        print(f"{'MULTIPLE CHOICE':^90}")
        line()
        for question, correct_answer in self.general.questions.items():# separates the question and answer
            answers = list(self.general.questions.values())
            answers.remove(correct_answer)  # Remove correct answer to avoid duplicate
            random.shuffle(answers)  # Shuffle remaining answers
            answers = answers[:3]  # Limit to three wrong answers
            answers.append(correct_answer)  # Add correct answer
            random.shuffle(answers) #Shuffle again with the correct answer
            for i in range(4): # Prints the choices
                  print(f"({choices[i]}) {answers[i]}")
            while True:
                answer = input(f"{question}: ").upper()
                if answer in choices:
                    for i in range(4):     #to get the answers base on the letter of the user's input
                      if answer == choices[i]:
                        self.general.answers.append(answers[i].upper()) # adds the answer to the general class
                    line()
                    break
                else:
                    line()
                    print("Invalid answer. Please try again.")
                    line()
        self.general.show_summary(type_of_questionnaire)
        line()
        print(f"{'YOU HAVE ANSWERED ALL OF THE QUESTIONS':^90}")
        line()

# VIEW CLASS ---------------------------------------------------------------------------
class View(General):
    def start(self):
        self.general = General("question_test.txt")  # Connecting the General instance
        self.general.read_file()  # Ensure questions are read from the file
        line()
        print(f"{'VIEW':^90}")
        line()
        print(f"{'Questions':<65}{'Answers':<20}")
        line()
        i=1
        for question, answer in self.general.dictionary.items():
            print(f"{f'({(i)}) {question}':<65}{answer:<20}")
            i += 1
            print()
        line()
        print(f"{'(a) Add':^25}{'(r) Remove':^25}{'(0) Back':^25}")
        line()
        while True:
            choice = input("Choose an option: ").lower()
            if choice == "a":
                self.add()
                break
            elif choice == "r":
                self.remove()
                break
            elif choice == "0":
                main()
                break
            else:
                print("Invalid option. Please Try Again")


    def remove(self):
      while True:
        num = input("Enter the number of the question to remove: ")
        if num.isdigit():
            num = int(num)
            if num in range(1, len(self.general.dictionary) + 1):  # Check valid number
                questions = list(self.general.dictionary.keys())  # Get list of questions
                question_to_remove = questions[num - 1]  # Get the question using the input number
                del self.general.dictionary[question_to_remove]  # Remove from the dictionary

                # Update the file after removing the question # rewrite
                with open("question_test.txt", "w") as file:
                    for question, answer in self.general.dictionary.items():
                        file.write(f"{question},{answer}\n")

                print("Question removed successfully!")
                self.start()
                break
            else:
                print("Question number not found in the dictionary. Please try again.")
        else:
            print("Invalid input. Please enter a number.")

    def add(self):
      line()
      add_question = input("Enter the question: ")
      line()
      add_answer = input("Enter the answer: ")
      line()
    # Add the new question and answer to the dictionary
      self.general.dictionary[add_question] = add_answer.upper()

    # Update the file with the new question
      with open("question_test.txt", "w") as file:
        for question, answer in self.general.dictionary.items():
            file.write(f"{question},{answer}\n")

      print("Question added successfully!")
      self.start()


#FUNCTIONS------------------------------------------------------------------------
def line(): # to create line
  print(f"{'-':-^90}")

def shuffle_questions(dictionary):  #shuffles the dictionary
    items = list(dictionary.items())  # Get all questions and answers
    random.shuffle(items)
    questions = dict(items)
    return questions

def show_type(): #show option for the type of questionnaire # Main Menu
  line()
  print(f"{'(1) '+ questionnaire_type[0]:^20} {'(2) '+ questionnaire_type[1]:^20}{'(0) Exit':^20}{'(v) View':^20}")
  line()
  while True:
    num = input("choose a questionnaire type: ").lower()
    if num == "1":
      choice = questionnaire_type[0]
      break
    elif num == "2":
      choice = questionnaire_type[1]
      break
    elif num == "0":
      choice = "0"
      break
    elif num == "v":
      choice = "v"
      break
    else:
      print("invalid option. Please Try Again")
      False
  return choice

#MAIN PROGRAM ------------------------------------------------------------------
def main():
    general = General("question_test.txt") #insert the filename path here/instance
    if general.read_file():
        line()
        print(f"{'QUIZ PROGRAM':^90}")
        line()
        choice = show_type() #show option type questionnaire
        if choice == questionnaire_type[0]:
            WordBank("question_test.txt").start(choice)
            choice = input("Do You Wish To Continue? (y to continue/anykey to exit): ").lower()
            if choice == "y":
                main()
            else:
                print("Exiting program.")
        elif choice == questionnaire_type[1]:
            if len(general.dictionary) >= 4:
                MultipleChoice("question_test.txt").start(choice)
                choice = input("Do You Wish To Continue? (y to continue/anykey to exit): ").lower()
                if choice == "y":
                    main()
                else:
                    print("Exiting program.")
            else:
                line()
                print("Not enough questions for multiple choice. Add more questions.")
                line()
                View("question_test.txt").start()

        elif choice == "0":
            print("Exiting program.")
        elif choice == "v":
            View("question_test.txt").start()
    else:
        print("Program will not commence due to file reading error.")

main()#COMMENCING THE PROGRAM------------------------------------------------------
```
